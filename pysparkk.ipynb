```python
# Step 1: Initialize Spark Session
spark = SparkSession.builder.appName("RDD_Pipeline").getOrCreate()
sc = spark.sparkContext  # Get Spark Context
```

```python
# Step 2: Sample Data
data = ["apple banana", "orange apple", "banana orange", "apple mango", "grape apple"]
```

```python
# Step 3: Create an RDD
rdd = sc.parallelize(data)
print("Original RDD:", rdd.collect())
```

```python
# Step 4: Transformation 1 - FlatMap (Splitting words)
rdd_flat = rdd.flatMap(lambda x: x.split())
print("After FlatMap:", rdd_flat.collect())
```

```python
# Step 5: Transformation 2 - Map (Convert words to lowercase)
rdd_lower = rdd_flat.map(lambda word: word.lower())
print("After Map (Lowercase):", rdd_lower.collect())
```

```python
# Step 6: Transformation 3 - Filter (Remove 'apple')
rdd_filtered = rdd_lower.filter(lambda word: word != "apple")
print("After Filter (Remove 'apple'):", rdd_filtered.collect())
```

```python
# Step 7: Transformation 4 - Distinct (Remove duplicates)
rdd_distinct = rdd_filtered.distinct()
print("After Distinct:", rdd_distinct.collect())
```

```python
# Step 8: Transformation 5 - SortBy (Sort words alphabetically)
rdd_sorted = rdd_distinct.sortBy(lambda word: word)
print("After SortBy (Alphabetical Order):", rdd_sorted.collect())
```

```python
# Step 9: Action - Collect results and print
result = rdd_sorted.collect()
print("Final Result:", result)
```

```python
# Step 10: Stop Spark session
spark.stop()
```
